```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'key': [1, 2, 4],
    'Marks': [90, 85, 95]
})
```
